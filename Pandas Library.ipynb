```python
# Austin temperature data
temperature_data = '2010 Austin weather.csv'

# read in the data set
austin_temp = pd.read_csv(temperature_data)

# check
austin_temp.head()
```

```python
# lets get only the temperature column for the first 800 rows
temp = austin_temp.iloc[:700, 0]

# Create a plot with color='red'
temp.plot(color = 'red')

# Add a plot title
plt.title('Austin Temperature')

# Specify the x-axis label
plt.xlabel('Hours since midnight August 1, 2010')

# Specify the y-axis label
plt.ylabel('Temperature (degrees F)')

# Display the plot
plt.show()
```

```python
# lets get only the temperature column
temp = austin_temp.iloc[:800, 0]
temp
```

```python
# Get just the numeric columns
austin_num = austin_temp.iloc[0:700, 0:3]

# check
austin_num.head()
```

```python
# plotting columns in all figure
austin_num.plot()
plt.show()
```

```python
# making separate subplots for column
austin_num.plot(subplots=True)
plt.show()
```

```python
# plot just one columns data
austin_num['DewPoint'].plot()
plt.show()
```

```python
# plot two columns data
austin_num[['Temperature', 'DewPoint']].plot()
plt.show()
```
